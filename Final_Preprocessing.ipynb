# <a name='Fun'>Imports Functions</a> 

```python
#This function shows the percentages of null values in each column
def get_numerical_summary(df):
    total = df.shape[0]
    missing_columns = [col for col in df.columns if df[col].isnull().sum() > 0]
    missing_percent = {}
    for col in missing_columns:
        null_count = df[col].isnull().sum()
        per = (null_count/total) * 100
        missing_percent[col] = per
        print("{} : {} ({}%)".format(col, null_count, round(per, 3)))
    return missing_percent
```

### Convert Placement_date, Harvest_Date and Slaughtering_Date columns from object to datetime

```python
df[["Placement_Date", "Harvest_Date", "Slaughtering_Date"]] = df[["Placement_Date", "Harvest_Date", 
                                                                  "Slaughtering_Date"]].apply(pd.to_datetime)
```

#### Converting object columns to integer

##### P_QTY, CUM, BAL, PDS, PDS_Mortality, P_ALW, Birds_Below_1_KG, HAN, DOA, Harvest_age_(If_harvested) and Variance columns are saved as object type, but they are integers.


```python
#P_QTY
#remove coma 
df['P_QTY'] = df['P_QTY'].str.replace(',', '') 
#convert object to int
df['P_QTY'] = df['P_QTY'].astype(str).astype(int)

#CUM
#remove coma 
df['Mortality'] = df['Mortality'].str.replace(',', '') 
#convert object to int
df['Mortality'] = df['Mortality'].astype(str).astype(int)

#BAL
#remove coma 
df['Harvest_QTY'] = df['Harvest_QTY'].str.replace(',', '') 
#convert object to int
df['Harvest_QTY'] = df['Harvest_QTY'].astype(str).astype(int)

#PDS
#remove coma 
df['PDS'] = df['PDS'].str.replace(',', '') 
#convert object to int
df['PDS'] = df['PDS'].astype(str).astype(int)

#PDS_Mortality
#remove () 
df['PDS_Mortality'] = df['PDS_Mortality'].str.extract(r'(\d+)', expand=False)
#remove coma 
df['PDS_Mortality'] = df['PDS_Mortality'].str.replace(',', '') 
#convert object to int
df['PDS_Mortality'] = df['PDS_Mortality'].astype(str).astype(int)


#P_ALW
#remove coma 
df['P_ALW'] = df['P_ALW'].str.replace(',', '') 
#convert object to int
df['P_ALW'] = df['P_ALW'].astype(str).astype(int)

#Birds_Below_1_KG
#remove coma 
df['Birds_Below_1_KG'] = df['Birds_Below_1_KG'].str.replace(',', '') 
#convert object to int
df['Birds_Below_1_KG'] = df['Birds_Below_1_KG'].astype(str).astype(int)

#HAN
#remove coma 
df['HAN'] = df['HAN'].str.replace(',', '') 
#convert object to int
df['HAN'] = df['HAN'].astype(str).astype(int)

#DOA
#remove coma 
df['DOA'] = df['DOA'].str.replace(',', '') 
#convert object to int
df['DOA'] = df['DOA'].astype(str).astype(int)

#Variance
#convert object to int
df['Variance'] = df['Variance'].astype(str).astype(int)

#Harvest_age_(If_harvested)
#convert object to int
df['Harvest_age_(If_harvested)'] = df['Harvest_age_(If_harvested)'].astype(float).astype(int)
```

#### Convert Placement_Density, Harvest_Density_/_Bird, Harvest_Density_/_KG, CV%, GB% and Dead_D_14 columns from object to float



```python
#Placement_Density
#convert object to float
df['Placement_Density'] = df['Placement_Density'].astype(str).astype(float)

#Harvest_Density_/_Bird
#convert object to float
df['Harvest_Density_/_Bird'] = df['Harvest_Density_/_Bird'].astype(str).astype(float)

#Harvest_Density_/_KG
#convert object to float
df['Harvest_Density_/_KG'] = df['Harvest_Density_/_KG'].astype(str).astype(float)

#Total_Area(sqm)
#convert object to float
df['Total_Area(sqm)'] = df['Total_Area(sqm)'].astype(str).astype(float)


#Dead_D_14
#convert object to float
df['Dead_D_14'] =df['Dead_D_14'].str.replace(' ', '')
df['Dead_D_14'] =df['Dead_D_14'].astype(float)


#CV
#remove %
df['CV%'] = df['CV%'].str.replace('%', '') 
#convert object to float
df['CV%'] = df['CV%'].astype(str).astype(float)


#GB
#remove %
df['GB%'] = df['GB%'].str.replace('%', '') 
#convert object to float
df['GB%'] = df['GB%'].astype(str).astype(float)


#Farm_M% 
#remove %
df['Farm_M%'] = df['Farm_M%'].str.replace('%', '') 
#convert object to float
df['Farm_M%'] = df['Farm_M%'].astype(str).astype(float)


#PDS_M% 
#remove %
df['PDS_M%'] = df['PDS_M%'].str.replace('%', '') 
#convert object to float
df['PDS_M%'] = df['PDS_M%'].astype(str).astype(float)
```
