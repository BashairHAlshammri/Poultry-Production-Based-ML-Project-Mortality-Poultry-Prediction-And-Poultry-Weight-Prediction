### In this plot we want to see the chicken quantity in summer vs winter 

```python
p_qty_df = dead_df.copy()
```

```python
p_qty_df .sort_values(by = 'PDS', ascending=False, inplace= True)
p_qty_df  = p_qty_df.reset_index()
```

```python
plt.figure(figsize=(10, 8))

sns.barplot(x='Season_of_harvest', y= 'PDS', data= p_qty_df, palette= ['#b1c9b6', '#5e8265'])
plt.title('Number Of Chickens vs Season',fontweight='bold')
plt.ylabel('Number Of Chickens',fontweight='bold')
plt.xlabel('Seasons',fontweight='bold')
plt.show()
```

 # <a name='pd4'>Fourth Plot </a>

### In this plot we want to see if the farms have an effect on the number of chickens that die on way (Dead On Arrival) 

```python
Farms_df = dead_df.copy()
```

```python
#Farms and DOA

Farms_chart = Farms_df.groupby('Farms')['DOA'].mean()

Farms_chart = Farms_chart.to_frame()
Farms_chart = Farms_chart.reset_index()
```

```python
# top 10 
Farms_chart.sort_values(by = 'DOA', ascending=False, inplace= True)
Farms_chart = Farms_chart.iloc[0:10]
```

```python
plt.figure(figsize=(10, 8))
current_palette = sns.color_palette("hls", 10)
sns.barplot(x='Farms', y= 'DOA', data= Farms_chart, palette = ['#b1c9b6', '#5e8265'])

plt.ylabel('Average Of Dead On Arrival',fontweight='bold')
plt.xlabel('Farm names',fontweight='bold')
plt.title('Farms and Average Dead On Arrival',fontweight='bold')
plt.show()
```

### In this plot we want to see what is the 5 most/least days that chickens die on

```python
sum_column = dead_day.sum(axis=0)
sum_column
```

```python
sum_column.sort_values()
```

```python
fig,df_dead_day = plt.subplots(figsize=(10, 5))
df_dead_day = sum_column.nlargest(n=5).plot(legend=False ,color="#5e8265",xlabel="Mortality days",ylabel="Number of birds" ).set_title("The highest 5 ranks of mortality days")
```

```python
fig,df_dead_day = plt.subplots(figsize=(10, 5))
df_dead_day = sum_column.nsmallest(n=5).plot(legend=False ,color="#5e8265" ,xlabel="Mortality days",ylabel="Number of birds" ).set_title("The least 5 ranks of mortality days")
```

```python
age_chart1 = weight_df.groupby("Harvest_age_(If_harvested)")['P_ALW'].mean()
age_chart = age_chart1.to_frame()
age_chart = age_chart.reset_index()
```

 # <a name='pw2'>Second Plot</a>

```python
plt.figure(figsize=(10, 7))
pie_chart = weight_df.groupby('Season_of_harvest')['Harvest_QTY'].mean()

pie_chart = pie_chart.to_frame()
pie_chart = pie_chart.reset_index()

Season = pie_chart['Season_of_harvest']
BAL_Avg = pie_chart['Harvest_QTY']

colors = ['#b1c9b6', '#5e8265']
plt.pie(BAL_Avg, labels = Season, autopct='%1.1f%%', colors=colors)
plt.title('Season of Harvest vs Number Of Chickens',fontweight='bold')

plt.show() 
```

```python
line_chart = weight_df.groupby('Month_of_harvest')['CV%'].mean()

line_chart = line_chart.to_frame()
line_chart = line_chart.reset_index()

Months = line_chart['Month_of_harvest']
PDS_Avg = line_chart['CV%']
fig, ax = plt.subplots(figsize=(8, 6))
tick_ls = ['Jan',"Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug"]
plt.xticks(range(0,8), tick_ls)
plt.xlabel('Months',fontweight='bold')
plt.ylabel('Coefficient Variation ',fontweight='bold')
plt.title('Coefficient Variation form Jan to Aug',fontweight='bold')
ax.plot(Months, PDS_Avg, color="#5e8265");
```
